###  TASK 2: Plot different lines for categories of vehicle type and analyse the trend to answer the question Is there a noticeable difference in sales trends between different vehicle types during recession periods?


<p>You will require:-
    <br>create a separate dataframe where the column recession has a value of '1'
    <br>to group the year, vehicle_type and calculate the average on the 'Automobile Sales'
    <br>one way is to -
    <br>use as_index as false else you will endup with multiple-indexed datafame
    <br>later set year as index and groupby vehicle over Sales and plot
    <br>make use of .plot() with kind = 'line'
    <br>do not forget to include labels and title
</p>


```python
# your code here
# Filter the data where Recession is 1
recession_data = df[df["Recession"] == 1]
# Group by Year and Vehicle_Type, calculating the mean Automobile Sales
answer = recession_data.groupby(["Year", "Vehicle_Type"], as_index=False)["Automobile_Sales"].mean()
## pivot(): This function reshapes the DataFrame, making the Vehicle_Type the columns and keeping Year as the index.
answer = answer.pivot(index="Year", columns="Vehicle_Type", values="Automobile_Sales")
# Fill missing data points with NaN using interpolation to make the lines continuous
answer = answer.interpolate(method="index", limit_area="inside")
# Plot the data
answer.plot(
    kind="line", 
    figsize=(10,6), 
    title="Average Automobile Sales during Recession by Vehicle Type",
    xlabel="Year",
    ylabel="Automobile _Sales"
)
plt.legend(title="Vehicle_Type", loc="center right")
```

### TASK 4: Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.
<br>Now, you want to find more insights from the data to understand the reason. <br>Plot a two line charts using subplotting to answer:-
#### How did the GDP vary over time during recession and non-recession periods? 
<br>Make use of <u>add_subplot()</u> from Matplotlib for this comparision.


```python
# your code here
# Separate data into recession and non-recession periods
recession_data = df[df["Recession"] == 1]
non_recession_data = df[df["Recession"] == 0]
# Group by Year and calc average GDP for each period
recession_gdp = recession_data.groupby("Year")["GDP"].mean()
non_recession_gdp = non_recession_data.groupby("Year")["GDP"].mean()

# Create subplots
fig, ax = plt.subplots(1, 2, figsize=(14, 6))
# Plot GDP during recession periods
ax[0].plot(recession_gdp.index, recession_gdp.values, label="GDP")
ax[0].set_title("Average GDP per Year in Recession Periods")
ax[0].set_xlabel("Year")
ax[0].set_ylabel("Average GDP")
ax[0].legend()
# Plot GDP during non-recession periods
ax[1].plot(non_recession_gdp.index, non_recession_gdp.values, label="GDP")
ax[1].set_title("Average GDP per Year in Non-Recession Periods")
ax[1].set_xlabel("Year")
ax[1].set_ylabel("Average GDP")
ax[1].legend()
# Set the main title
fig.suptitle("Comparing GDP in Recession and Non-Recession Periods")
```

### TASK 5: Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend

##### Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>
Title this plot as 'Seasonality impact on Automobile Sales'


<p>You can create Bubble Chart by calling the scatter()
        <br>Pass the 'Month' and 'Automobile_Sales' to the functions as x and y and then use Seasonality weight for size parameter</p>


```python
# your code here
month_order = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
# Using non-recession data
# Group the data by month and calc the mean Automobile Sales
non_recession_data = df[df["Recession"] == 0]
monthly_sales = non_recession_data.groupby("Month")["Automobile_Sales"].mean().reset_index()
# Merge the seasonality weight back to the monthly_sales data 
monthly_sales = monthly_sales.merge(df[["Month", "Seasonality_Weight"]].drop_duplicates(), on="Month", how="left")
# Sort "Month" column is a categorical type with the specified order
monthly_sales["Month"] = pd.Categorical(monthly_sales["Month"], categories=month_order, ordered=True)
monthly_sales = monthly_sales.sort_values("Month")

plt.figure(figsize=(10, 6))
# Creating scatter plot
plt.scatter(
    x=monthly_sales["Month"],
    y=monthly_sales["Automobile_Sales"],
    s=monthly_sales["Seasonality_Weight"] * 100,  # Bubble size
    color="blue"
)
# Add the title, labels and legend
plt.title("Seasonality impact on Automobile Sales during Non-Recession Years")
plt.xlabel("Month")
plt.ylabel("Automobile_Sales")
# Legend part
sizes = [0.25, 0.50, 0.75, 1.00, 1.25, 1.50]
scatter_sizes = [s * 100 for s in sizes]
for size in scatter_sizes:
    plt.scatter([], [], s=size, color="blue", label=str(size / 100))
plt.legend(title="Seasonality_Weight", loc="lower right")
```

## Swiss Banknote Dataset

Six variables measured on 100 genuine and 100 counterfeit old Swiss 1000-franc
bank notes. The data stem from Flury and Riedwyl (1988). The columns correspond
to the following one label and six variables:

0. Conterfeit - 0: genuine bank notes„ÄÅ1: genuine bank notes (Labels)


1. Length - Length of the bank note


2. Left - Height of the bank note, measured on the left


3. Right - Height of the bank note, measured on the right


4. Bottom - Distance of inner frame to the lower border


5. Top - Distance of inner frame to the upper border


6. Diagonal - Length of the diagonal



More details: [Kaggle | Swiss banknote conterfeit detection](https://www.kaggle.com/chrizzles/swiss-banknote-conterfeit-detection)

```python
# Use [Pandas] to import data | banknotes.csv
# Put your code here 

bank = pd.read_csv("banknotes.csv")
label = bank.conterfeit

bank.head()
```

```python
# Use [Numpy function] to stack the following columns
# The 5-th column "Bottom" as X, the 6-th column "Top" as Y
# Put your code here
#data_xy = np.stack((bank.Bottom, bank.Top), axis=1)    # same
data_xy = np.stack((bank["Bottom"], bank["Top"]), axis=1)
data_xy
```

```python
# Use [Numpy function] to find where label = 0 / label = 1
# Put your code here

index_0 = np.where(label == 0)  # Real banknotes
index_1 = np.where(label == 1)  # Fake banknotes


# Use data_xy to plot the scatter plot 
# Label = 0 ‚Üí color = green | Label = 1 ‚Üí color = red
# Fill X and Y in the plt.scatter 
# Remember to put on grid and legend and equal axis
# Put your code here

plt.figure(figsize = (10, 10))
plt.scatter(data_xy[index_0, 0], data_xy[index_0, 1], color="green", label="real banknote", s=20, alpha=0.8)
plt.scatter(data_xy[index_1, 0], data_xy[index_1, 1], color="red", label="fake banknote", s=20, alpha=0.8)

# Add the grid, legend, and equal axis 
plt.grid(True)
plt.legend(loc="upper right")
plt.axis("equal")
```

```python
# Use [Numpy function] to calculate the squared distance matrix between each points
# Put your code here
# Reshapes data_xy so that it broadcasts correctly for pairwise distance calculations. 
# It adds a new axis to create a 3D array where each point is compared against all other points.
dist_sq = np.sum(np.square(data_xy[:, np.newaxis] - data_xy[np.newaxis, :]), axis=2)
```

```python
dist_sq
```

```python
def nearest_partition(dist_sq, K):
    
    # Use [Numpy function] to return the "K Nearest Neighbor"
    # Put your code here
    # For each point, find the indices of the K nearest neighbors
    # Set the diagonal (self-distance) to infinity to exclude the point itself
    np.fill_diagonal(dist_sq, np.inf)
    # Find the K nearest neighbors, excluding the point itself
    nearest_neighbors = np.argpartition(dist_sq, K, axis=1)[:, : K]
    
    return nearest_neighbors
```

```python
# Use [Numpy function] to find where label = 0 / label = 1
# Put your code here

index_0 = np.where(label == 0)
index_1 = np.where(label == 1)


# Use data_xy to plot the scatter plot 
# Label = 0 ‚Üí color = green | Label = 1 ‚Üí color = red
# Fill X and Y in the plt.scatter
# Remember to put on grid and legend and equal axis
# Put your code here

plt.figure(figsize = (10, 10))
plt.scatter(data_xy[index_0, 0], data_xy[index_0, 1], color="green", label="real banknote", s=20, alpha=0.8)
plt.scatter(data_xy[index_1, 0], data_xy[index_1, 1], color="red", label="fake banknote", s=20, alpha=0.8)

# Add the grid, legend, and equal axis 
plt.grid(True)
plt.legend(loc="upper right")
plt.axis("equal")

        
# Draw lines from each point to its three nearest neighbors (set K=3)
# Use some zip magic to make it happen (Hint is at below cell)
# You might need loops to generate the plot below
# set parameter [color='black', linewidth=1, alpha=0.5] when you draw lines 
# Put your code here

K = 3
# Draw the lines between each point and its 3 nearest neighbors
# I guess no need to use zip magic
nearest_neighbors = nearest_partition(dist_sq=dist_sq, K=K)
for i, neighbors in enumerate(nearest_neighbors):
    for neighbor in neighbors:
        plt.plot([data_xy[i, 0], data_xy[neighbor, 0]], [data_xy[i, 1], data_xy[neighbor, 1]], color="black", linewidth=1, alpha=0.5)

```

```python
# Set seed = 10 | Random select three dots from data_xy
# Save dots index in varible "target" 
# Don't need to change below three lines

random.seed(10)
target = np.array([random.randint(0, len(data_xy)) for i in range(3)])
text = ['A', 'B', 'C']


# Use [Numpy function] to find where label = 0 / label = 1
# Put your code here

index_0 = np.where(label == 0)
index_1 = np.where(label == 1)


# Use data_xy to plot the scatter plot 
# Label = 0 ‚Üí color = green | Label = 1 ‚Üí color = red | Target ‚Üí color = blue
# Fill X and Y in the plt.scatter
# Remember to put on grid and legend and equal axis
# Put your code here

plt.figure(figsize = (12, 12))
plt.scatter(data_xy[index_0, 0], data_xy[index_0, 1], color="green", label="real banknote", s=20, alpha=0.8)
plt.scatter(data_xy[index_1, 0], data_xy[index_1, 1], color="red", label="fake banknote", s=20, alpha=0.8)
plt.scatter(data_xy[target, 0], data_xy[target, 1], color="blue", label="target dots", s=60)    # Target dots

# Add the grid, legend, and equal axis
plt.grid(True)
plt.legend(loc="upper right")
plt.axis("equal")

# Draw lines from the selected three point to its three nearest neighbors (set K=3)
# Use some zip magic to make it happen (Hint is at below cell)
# You might need loops to generate the plot below
# set parameter [color='black', linewidth=1, alpha=0.5] when you draw lines 
# Put text A, B, C on each dots (location: Right + 0.1„ÄÅ Up + 0.1)
# Put your code here


K = 3
nearest_neighbors = nearest_partition(dist_sq=dist_sq, K=K)
for point_random in target:
    for neighbor in nearest_neighbors[point_random]:
        #print(f"Neighbors for point A (index {target[0]}): {nearest_neighbors[target[0]]}")
        plt.plot([data_xy[point_random, 0], data_xy[neighbor, 0]], 
                 [data_xy[point_random, 1], data_xy[neighbor, 1]], 
                 color='black', linewidth=1, alpha=0.5)

# Put text A, B, C (location: Right + 0.1, Up + 0.1)
for i, point in enumerate(target):
    plt.text(data_xy[point, 0] + 0.1, data_xy[point, 1] + 0.1, text[i], fontsize=12)

```
